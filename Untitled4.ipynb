```python
temperature_df = pd.DataFrame({
    "city" : ["Rouma","Poori","Helsinki"],
    "temperature" : [26,23,30]
})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city" : ["Helsinki","Rouma","Poori"],
    "humidity" : [56,45,33]
})
humidity_df
```

```python
df_3 = pd.concat([temperature_df,humidity_df], axis = 1)
df_3
```

```python
temperature_df.set_index('city', inplace=True)
humidity_df.set_index('city', inplace=True)

# Concatenate the DataFrames along the columns axis
df_4 = pd.concat([temperature_df, humidity_df], axis=1)

# Reset the index if you want 'city' as a column again
df_4.reset_index(inplace=True)

print(df_4)
```

```python
df_5 = pd.Series(["Humid","Dry","Rain"],name = "event")
df_5
```

```python
df_6 =pd.concat([df_4,df_5],axis = 1)
df_6
```
