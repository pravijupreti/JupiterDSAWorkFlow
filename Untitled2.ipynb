```python
import chardet 

def detect_encoding(file_path):
    with open(file_path,'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

file_path = "Employee Sample Data.csv"
encoding =  detect_encoding(file_path)
print(f"Detected encoding{encoding}")
try : 
    df_3 = pd.read_csv(file_path,encoding = encoding)
    print("CSV file read successfully")
except UnicodeDecodeError as e:
    print(f"Error reading the CSV file{e}")

df_filled = df_3.fillna("n.a.")
print("DAta after replacing missing value with 'n.a.':")
print(df_filled)

df_filled.to_csv("Employee_Sample_Data_Cleaned.csv", index=False)
```
