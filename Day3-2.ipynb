```python
df1 = pd.DataFrame({
    "city" : ["Helsinki","Espoo","Pori","Rouma","Vanta"],
    "temperature" : [20,22,21,24,19]
})
df1
```

```python
df2 = pd.DataFrame({
    "city" : ["Helsinki","Espoo","Vanta","Porvo"],
    "Humidity" : [46,56,48,57]
})
df2
```

```python
df_outer = pd.merge(df1,df2,on="city",how="outer")
df_outer
```

```python
df_inner = pd.merge(df1,df2,on="city",how="inner")
df_inner
```

```python
df_right = pd.merge(df1,df2,on="city",how="right")
df_right
```

```python
df_left = pd.merge(df1,df2,on="city",how="left")
df_left
```

```python
df_pivot = df_outer.pivot(index = "city",columns = "temperature",values = "Humidity")
df_pivot
```

```python
df_8 = pd.DataFrame({
    "Name" :["Kathy","Linda","Peter","John","Pravij","Babita","Prakash","Laxmi","Raj","Pankaj","Yan","Juan"],
    "Nationality" :["Finland","Finland","Finland","Finland","Nepal","Nepal","Nepal","Nepal","India","India","Chaina","Chaina"],
    "Sex" : ["Female","Female","Male","Male","Male","Female","Male","Female","Male","Male","Male","Female"],
    "Age" : [18,19,22,31,35,24,21,23,32,23,23,22],
    "Handedness" : ["Right","Right","Left","Right","Left","Right","Left","Right","Right","Left","Right","Right"]
})
df_8
```

```python
pd1_8 = pd.crosstab(df_8.Nationality,df_8.Handedness)
pd1_8
```

```python
pd2_8 = pd.crosstab(df_8.Sex,df_8.Handedness,margins = True)
pd2_8
```

```python
pd3_8 = pd.crosstab([df_8.Sex,df_8.Age],[df_8.Handedness,df_8.Nationality],margins = True)
pd3_8
```

```python
pd4_8 = pd.crosstab(df_8.Sex,df_8.Handedness,normalize ='index')
pd4_8
```

```python
pd5_8 = pd.crosstab([df_8.Sex],[df_8.Handedness],values = df_8.Age,aggfunc = np.average)
pd5_8
```
